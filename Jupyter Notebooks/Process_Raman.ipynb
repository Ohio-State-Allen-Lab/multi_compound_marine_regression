```python
#Data Processing and Management
import os
import glob
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Path for the Raman Data
n_raman_path = r"C:\Users\nmbis\OneDrive - The Ohio State University\Desktop\Current Work\Papers\Ocean ML Paper\Experimental_Data\N_Files"
# Find all Raman Files
os.chdir(n_raman_path)
n_raman_all_files = glob.glob(n_raman_path + "/*.csv")
```

# Spiked Samples With Plain Water

```python
N_Raman_Array = pd.DataFrame()
for filename in n_raman_all_files:
    #Need to separate the CAS from the filename
    path= r"C:\Users\nmbis\OneDrive - The Ohio State University\Desktop\Current Work\Papers\Ocean ML Paper\Experimental_Data\N_Files"
    f_name = filename.replace(path,'')
    f_name = f_name.replace('\\','')
    f_name = f_name.replace('.csv','')
    split_name = f_name.split('  ', 1)
    f_name = split_name[0]
    #print(f_name)
    df = pd.read_csv(filename, index_col=False, header=0)
    #if "Wavelength" not in Raman_Array:
    #    Raman_Array.insert(0, 'Wavelength', df.iloc[:, 0])
    if f_name not in N_Raman_Array:
        temp_df = pd.DataFrame({f_name : df.iloc[:, 1]})
        N_Raman_Array = pd.concat((N_Raman_Array, temp_df), axis = 1)
N_Raman_Array.to_csv(r'C:\Users\nmbis\OneDrive - The Ohio State University\Desktop\Current Work\Papers\Ocean ML Paper\N_Dataset.csv', index = False, encoding='utf-8')
```
